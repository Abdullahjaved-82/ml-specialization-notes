## Vectors

### Vector Creation
NumPy gives multiple ways to create vectors — some fill with default values, some with random values, some with your own specific numbers.

```python
a = np.zeros(4)
print(f"a: {a}, size: {a.size}, datatype: {a.dtype}") 
a = np.zeros((4,))
print(f"a: {a}, size: {a.size}, datatype: {a.dtype}")
a = np.random.random_sample(4)
print(f"a: {a}, size: {a.size}, datatype: {a.dtype}")
```

Some data creation routines do not take a shape tuple as input.

```python
a = np.arange(4.)
print(f"np.arange(4.): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")

a = np.arange(0,30,3)
print(f"np.arange(4.): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")

a = np.random.rand(4)
print(f"np.random.rand(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
```

```python
a = np.array([5,4,3,2])
print(f"np.array([5,4,3,2]) {a} , shape: {a.shape} , Datatype: {a.dtype}")
a = np.array([5.,4,3,2])
print(f"np.array([5.,4,3,2]) {a} , shape: {a.shape} , Datatype: {a.dtype}")
```

### Indexing
Indexing means referring to a single element of an array by its position.

```python
a = np.arange(10)
print(a)
print(f"a[2] {a[2]}")
print(f"a[-1] {a[-1]}")
print(f"a[-2] {a[-2]}")
```

### Slicing
Slicing creates a subset of an array using start:stop:step, similar to how arange works, but applied to an existing array instead of generating new numbers.

```python
a = np.arange(10)
print(f"a {a}")

c = a[2:8:1]
print(f"c {c}")
#IN slicing ,  the position of each number and its value are exactly the same number. Position 2 holds the value 2
c = a[2:7:2]
print(f"c {c}")

c = a[3:]
print(f"c {c}")

c = a[:7]
print(f"c {c}")

c = a[:]
print(f"c {c}")
```

### Single Vector Operations
Operations you can apply to just one vector at a time — negation, sum, mean, and squaring.

```python
a = np.array([1,3,5,7,9])
print(f"a {a}")

b = -a
print(f"-a {b}")

b = np.sum(a)
print(f"b = np.sum(a) : {b}")

b = np.mean(a)
print(f"b = np.mean(a): {b}")

b = a**2
print(f"b = a**2 : {b}")


```

Operations between two vectors, applied position by position.

```python
a = np.array([1,2,3,4])
b = np.array([-1,-2,-3,4])

print(f"binary operators works element wise  {a+b}")

c = np.array([1,2])
try:
    d = c+a
except Exception as e:
    print(f"the error message you see is {e}")
```

### Scalar Vector Operations
Multiplying an entire vector by a single number.

```python
a = np.array([1,2,3,4])
b = 5*a
print(f"b {b}")
```

### The Dot Product
The dot product multiplies two vectors element-wise, then sums the results into a single number. It's one of the most important operations in this entire course.

```python
def my_dot(a,b):
    x =0
    for i in range(a.shape[0]):
        x = x+a[i]*b[i]
    return x

a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
print(f"my_dot(a, b) = {my_dot(a, b)}")
```

Let's compare our own version against NumPy's built-in np.dot function.

```python
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])

c = np.dot(a, b)
print(f"NumPy 1-D np.dot(a, b) = {c}, np.dot(a, b).shape = {c.shape}")

c = np.dot(b, a)
print(f"NumPy 1-D np.dot(b, a) = {c}, np.dot(a, b).shape = {c.shape}")
```

### The Need for Speed
NumPy is faster than plain Python loops because it uses vectorized operations under the hood. Let's prove it by timing both approaches on a huge array.

```python
# np.random.seed(1) locks the random number generator to a fixed starting point,
# so the "random" numbers below are the same every time this cell runs.
# The number 1 has no special meaning - any number would work the same way,
# it's just used here so results are repeatable and match the lab example.

np.random.seed(1)
a = np.random.rand(100000)
b = np.random.rand(100000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()

print(f"np.dot(a, b) = {c:.4f}")
print(f"Vectorized version duration: {1000*(toc-tic):.4f} ms")

tic = time.time()
c = my_dot(a, b)
toc = time.time()



print(f"my_dot(a, b) = {c:.4f}")
print(f"loop version duration: {1000*(toc-tic):.4f} ms")

del(a)
del(b)
```

## Matrices
Matrices are two-dimensional arrays — think of them as a grid of numbers, with rows and columns.

```python
a = np.array([
 [1000,5,10],
 [1500,3,5],
[1200,4,8]
])
```

### Matrix Creation
The same functions that create vectors also create matrices — you just give them a 2D shape instead of a single number.

```python
a = np.zeros((1, 5))
print(f"a shape = {a.shape}, a = {a}")

a = np.zeros((2, 1))
print(f"a shape = {a.shape}, a = {a}")

a = np.random.random_sample((1, 1))
print(f"a shape = {a.shape}, a = {a}")
```

```python
a = np.array([[5], [4], [3]])
print(f"a shape = {a.shape}, np.array: a = {a}")

a = np.array([[5],
              [4],
              [3]])
print(f"a shape = {a.shape}, np.array: a = {a}")
```

### Matrix Indexing
Matrices need two indexes to access an element: [row, column].

```python
a = np.arange(6).reshape(-1, 2)
print(f"a.shape: {a.shape}, \na= {a}")

print(f"\na[2,0].shape: {a[2, 0].shape}, a[2,0] = {a[2, 0]}, type(a[2,0]) = {type(a[2, 0])}")

print(f"a[2].shape: {a[2].shape}, a[2] = {a[2]}, type(a[2]) = {type(a[2])}")
```

### Matrix Slicing
Slicing works on matrices too, using the same start:stop:step pattern, but now for both rows and columns.

```python
# Create a matrix: 20 numbers reshaped into 2 rows, 10 columns each
a = np.arange(20).reshape(-1, 10)
print(f"a = \n{a}")

# Row 0 only, then slice columns 2 through 6 within that row -> returns a 1-D vector
print("a[0, 2:7:1] = ", a[0, 2:7:1], ", a[0, 2:7:1].shape =", a[0, 2:7:1].shape, "a 1-D array")

# All rows (the bare ':'), but only columns 2 through 6 -> returns a 2-D array
print("a[:, 2:7:1] = \n", a[:, 2:7:1], ", a[:, 2:7:1].shape =", a[:, 2:7:1].shape, "a 2-D array")

# Bare ':' for both rows and columns -> returns the entire matrix, unchanged
print("a[:,:] = \n", a[:,:], ", a[:,:].shape =", a[:,:].shape)

# Row 1, all columns, written explicitly -> returns a 1-D vector
print("a[1,:] = ", a[1,:], ", a[1,:].shape =", a[1,:].shape, "a 1-D array")

# Row 1, written without the column part -> same result as above, shorter syntax
print("a[1] = ", a[1], ", a[1].shape =", a[1].shape, "a 1-D array")
```
